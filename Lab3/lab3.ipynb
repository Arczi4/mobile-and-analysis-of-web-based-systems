```python
df_upload = pd.read_csv(os.path.join('data', 'merged_upload.csv'))
```

```python
df_upload['speed'].groupby(df_upload['operator']).mean().plot(kind='bar', title='Mean upload speed')
```

```python
import seaborn as sns
import numpy as np
# Correlation matrix
corrmat = df_upload.corr()

# Saleprice correlation matrix
k = 10 # numbers of variables in heatmap
cols = corrmat.nlargest(k, 'speed')['speed'].index
print(cols)
f, ax = plt.subplots(figsize=(10, 10))
cm = np.corrcoef(df_upload[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
```
